```python
# Librerias de ayuda
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


print(tf.__version__)
```

```python
# Establecer la semilla para reproducibilidad
np.random.seed(0)

# Generar datos
horas_estudio = np.random.rand(100, 1) * 10  # Horas de estudio entre 0 y 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2  # Calificaciones
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
```

```python
# Crear el modelo de regresión lineal
modelo = LinearRegression()
```

```python
# Entrenar el modelo con los datos de entrenamiento
modelo.fit(train_data, train_labels)
```

```python
# Realizar predicciones con el conjunto de datos de prueba
predictions = modelo.predict(test_data)
```

```python
# Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predictions)
print("Error cuadrático medio (MSE):", mse)
```

```python
# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(horas_estudio, calificaciones, color='blue', label='Datos originales')
plt.scatter(test_data, predictions, color='red', label='Predicciones')
plt.plot(test_data, predictions, color='black', label='Línea de regresión')
plt.title('Regresión Lineal - Predicción de Calificaciones')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.legend()
plt.show()
```
